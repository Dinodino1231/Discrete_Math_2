Implementation of CODES:

$\textbf{Kruskal's Algorithm}$

```python
import matplotlib.pyplot as plt
import networkx as nx

def kruskal_mst(graph):
    mst = nx.Graph()
    mst.add_nodes_from(graph.nodes)
    
    # Sort edges by weight
    sorted_edges = sorted(graph.edges(data=True), key=lambda x: x[2]['weight'])
    
    # Union-Find data structure to detect cycles
    parent = {node: node for node in graph.nodes}
    rank = {node: 0 for node in graph.nodes}
    
    def find(node):
        if parent[node] != node:
            parent[node] = find(parent[node])
        return parent[node]
    
    def union(node1, node2):
        root1 = find(node1)
        root2 = find(node2)
        if root1 != root2:
            if rank[root1] > rank[root2]:
                parent[root2] = root1
            else:
                parent[root1] = root2
                if rank[root1] == rank[root2]:
                    rank[root2] += 1
    
    # Iterate through sorted edges and add to MST if it doesn't form a cycle
    for u, v, data in sorted_edges:
        if find(u) != find(v):
            mst.add_edge(u, v, weight=data['weight'])
            union(u, v)
    
    return mst

seed = 55
# Use seed for reproducibility
G = nx.Graph()
plt.figure(figsize=(10, 8))
# Add nodes
G.add_nodes_from(["A","B","C","D","E","F","G"])
# Add self-loops to the remaining nodes
edgelist = [("A","C"),("A","I"),("C","B"),("C","E"),("C","D"),("B","E"),("I","E"),("E","F"),("F","G"),("G","E")]
G.add_edges_from(edgelist)

# Add edge weights
edge_weights = {("A","C"):10,("A","I"):5,("C","B"):20,("C","E"):15,("C","D"):30,("B","E"):25,("I","E"):40,("E","F"):35,("F","G"):50,("G","E"):45}
nx.set_edge_attributes(G, edge_weights, 'weight')

# Compute MST using Kruskal's algorithm
mst = kruskal_mst(G)

# Draw original graph
pos = nx.planar_layout(G)  # Seed for reproducible layout
nx.draw(G, pos=pos, with_labels=True)
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_weights, font_color='red')

# Draw MST
nx.draw_networkx_edges(mst, pos=pos, edge_color='blue', width=2)

plt.show()
```

$\textbf{Prim's Algorithm}$

```python
import matplotlib.pyplot as plt
import networkx as nx

def prim_mst(graph):
    mst = nx.Graph()
    mst.add_node(list(graph.nodes)[0])  # Start with an arbitrary node
    
    while len(mst) < len(graph.nodes):
        min_edge = None
        min_weight = float('inf')
        
        for u in mst.nodes:
            for v in graph.neighbors(u):
                if v not in mst.nodes:
                    weight = graph[u][v]['weight']
                    if weight < min_weight:
                        min_weight = weight
                        min_edge = (u, v)
        
        if min_edge:
            u, v = min_edge
            mst.add_edge(u, v, weight=min_weight)
    
    return mst

seed = 55
# Use seed for reproducibility
G = nx.Graph()
plt.figure(figsize=(10, 8))
# Add nodes
G.add_nodes_from(["A","B","C","D","E","F","G"])
# Add self-loops to the remaining nodes
edgelist = [("A","C"),("A","I"),("C","B"),("C","E"),("C","D"),("B","E"),("I","E"),("E","F"),("F","G"),("G","E")]
G.add_edges_from(edgelist)

# Add edge weights
edge_weights = {("A","C"):10,("A","I"):5,("C","B"):20,("C","E"):15,("C","D"):30,("B","E"):25,("I","E"):40,("E","F"):35,("F","G"):50,("G","E"):45}
nx.set_edge_attributes(G, edge_weights, 'weight')

# Compute MST using Prim's algorithm
mst = prim_mst(G)

# Draw original graph
pos = nx.planar_layout(G)  # Seed for reproducible layout
nx.draw(G, pos=pos, with_labels=True)
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_weights, font_color='red')

# Draw MST
nx.draw_networkx_edges(mst, pos=pos, edge_color='blue', width=2)

plt.show()
```

$\textbf{Djitraka's Shortest Path Algorithm}$

```python
import matplotlib.pyplot as plt
import networkx as nx

def dijkstra_shortest_path(G, start):
    visited = {start: 0}
    path = {}

    nodes = set(G.nodes)

    while nodes:
        min_node = None
        for node in nodes:
            if node in visited:
                if min_node is None:
                    min_node = node
                elif visited[node] < visited[min_node]:
                    min_node = node

        if min_node is None:
            break

        nodes.remove(min_node)
        current_weight = visited[min_node]

        for neighbor in G.neighbors(min_node):
            weight = current_weight + G[min_node][neighbor].get("weight", 1)
            if neighbor not in visited or weight < visited[neighbor]:
                visited[neighbor] = weight
                path[neighbor] = min_node

    return visited, path

seed = 55
G = nx.Graph()
plt.figure(figsize=(10, 8))

G.add_nodes_from(["A", "B", "C", "D", "E", "F", "G"])
edgelist = [("A", "C"), ("A", "I"), ("C", "B"), ("C", "E"), ("C", "D"), ("B", "E"), ("I", "E"), ("E", "F"), ("F", "G"), ("G", "E")]

G.add_edges_from(edgelist)

# Set edge weights
edge_weights = {("A", "C"): 10, ("A", "I"): 5, ("C", "B"): 20, ("C", "E"): 15, ("C", "D"): 30, ("B", "E"): 25, ("I", "E"): 40, ("E", "F"): 35, ("F", "G"): 50, ("G", "E"): 45}
nx.set_edge_attributes(G, edge_weights, "weight")

pos = nx.planar_layout(G)
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_weights, font_color='red')

# Highlight visited edges and nodes in blue
visited_nodes = ['A', 'C', 'B', 'E', 'D', 'F', 'G', 'I']
nx.draw(G, pos=pos, with_labels=True, node_color='blue', nodelist=visited_nodes, node_size=500, alpha=0.9, edge_color='blue', edgelist=edgelist, width=2, linewidths=2)

# Running Djikstra's Algorithm
start_node = "A"
distances, shortest_path_tree = dijkstra_shortest_path(G, start_node)
print("Shortest distances from node", start_node + ":")
print(distances)

plt.show()
```
