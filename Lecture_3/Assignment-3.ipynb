$\textbf{SIMPLE GRAPHS}$

 A simple graph is a graph that does not have more than one edge between any two vertices and no edge starts and ends at the same vertex. In other words a simple graph is a graph without loops and multiple edges.

 Definition of a simple graph would be:

 - No multiple edges
 - No self-loops
 - Pair of vertices
 - Undirected
 - Unweighted
 - Finite

```python
import matplotlib.pyplot as plt
import networkx as nx
seed = 24
# Use seed for reproducibility
G = nx.Graph()

# Add nodes
G.add_nodes_from([1,2,3,4])
# Add self-loops to the remaining nodes
edgelist = [(2,4),(1,3)]
G.add_edges_from(edgelist)
#pos = nx.spring_layout(G, seed=seed)  # Seed for reproducible layout
pos = nx.planar_layout(G)
nx.draw(G, pos=pos,with_labels=True)
plt.show()
```

a $\textbf{complete graph}$ is a simple undirected graph in which every pair of distinct vertices is connected by a unique edge.

```python
import matplotlib.pyplot as plt
import networkx as nx

# Use seed for reproducibility
seed = 24
G = nx.Graph()

# Add nodes
G.add_nodes_from([0, 1, 2, 3, 4])

# Add edges to create a complete graph
edgelist = [(i, j) for i in range(5) for j in range(i+1, 5)]
G.add_edges_from(edgelist)

# Use spring layout for better visualization of a complete graph
pos = nx.spring_layout(G, seed=seed)

# Draw the graph
nx.draw(G, pos=pos, with_labels=True)
plt.show()
```

$\textbf{BIPARTITE GRAPH}$

A bipartite graph is a graph whose vertices can be divided into two disjoint and independent sets and, that is, every edge connects a vertex in to one in.

$\textbf{Characteristics:}$

Vertex Partition: The vertices of a bipartite graph can be partitioned into two disjoint sets $U$and $V$ such that every edge connects a vertex from $U$ to a vertex from $V$.

Edge Set: All edges in a bipartite graph connect a vertex from set $U$ to a vertex from set $V$. There are no edges connecting vertices within the same set.

No Odd Cycles: Bipartite graphs do not contain any odd cycles. This property distinguishes them from non-bipartite graphs, which can contain odd cycles. A cycle with an odd number of vertices is called an odd cycle.

```python
import matplotlib.pyplot as plt
import networkx as nx

# Create an empty bipartite graph
G = nx.Graph()

# Define the sets of nodes for the bipartite graph
set_U = [1, 2, 3, 4]
set_V = ['A', 'B', 'C']

# Add nodes from set U and set V to the graph
G.add_nodes_from(set_U, bipartite=0)  # Mark nodes in set U as 'bipartite=0'
G.add_nodes_from(set_V, bipartite=1)  # Mark nodes in set V as 'bipartite=1'

# Define edges between nodes in set U and set V
edges = [(1, 'A'), (1, 'B'), (1, 'C'), (2, 'A'), (2, 'B'), (2, 'C'), (3, 'A'), (3, 'B'), (3, 'C'), (4, 'A'), (4, 'B'), (4, 'C')]
G.add_edges_from(edges)

# Draw the bipartite graph
pos = nx.bipartite_layout(G, set_U)
nx.draw(G, pos=pos, with_labels=True, node_size=500)
plt.title('Bipartite Graph')
plt.show()
```

The graph above can be considered as bipartite graph as the vertices can be partitioned into two disjoint sets, there are no edges connecting vertices within the same set, and does not contain an odd cycle. Therefore, the graph shows as bipartite graph.

$\textbf{TRAVELLING IN A GRAPH}$

Travelling in a graph typically refers to the process of moving from one vertex to another along the edges of the graph.

- $Walk$: A walk in a graph is a sequence of vertices and edges that alternately follow vertices and edges, starting and ending at vertices. It can visit the same vertex or edge multiple times.

- $Trail$: A trail in a graph is a walk in which no edge is repeated (but vertices may be repeated). In other words, a trail is a sequence of vertices and edges that do not repeat edges.

- $Path$: A path in a graph is a walk in which no vertex or edge is repeated (except that the first and last vertices may be the same, forming a cycle). In other words, a path is a sequence of vertices and edges that do not repeat.

- $Closed\ Walk$: A closed walk in a graph is a walk that starts and ends at the same vertex, possibly repeating vertices and edges.

- $Circuit$: A circuit in a graph is a closed walk that starts and ends at the same vertex, without repeating any vertices or edges (except for the starting/ending vertex). In other words, it is a path that forms a closed loop.

- $Simple\ Circuit$: A simple circuit in a graph is a closed walk in which no vertex (except the starting/ending vertex) is repeated, and no edge is repeated (except for the starting/ending edge).

$\textbf{CONNECTED GRAPH}$

A connected graph is a graph in which there is a path between every pair of vertices. In other words, a connected graph cannot be split into two or more separate subgraphs where there is no path between the vertices in different subgraphs.

$Characteristic:$

- Every vertex is reachable from every other vertex via a sequence of edges.
- There are no isolated vertices (vertices that are not connected to any other vertex).
- There are no disconnected components within the graph.

```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a connected graph
G = nx.Graph()

# Add edges to create a connected graph
G.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 5), (5, 1)])

# Draw the graph
pos = nx.spring_layout(G)  # Position nodes using the spring layout algorithm
nx.draw(G, pos, with_labels=True, node_size=700, font_size=12, font_weight="bold")
plt.show()
```

The graph above shows a connected graph, as every vertex is connected to another one, there are no isolated vertex, and there are no disconnected components.

$\textbf{CONNECTED COMPONENTS}$

Connected components are subsets of the vertices in a graph where every vertex in the subset is connected to every other vertex in the subset by paths, and no vertex in the subset is connected to any vertex outside the subset. In simpler terms, a connected component is a maximal subgraph in which all vertices are connected to each other, and there are no connections to vertices outside the subgraph.

$\textbf{Characteristics:}$

- Every vertex in $V'$ is reachable from every other vertex in $V'$ via a path in $G$.
- No vertex in $V'$ is connected to any vertex outside of $V'$.

```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a graph with multiple connected components
G = nx.Graph()
G.add_edges_from([(1, 2), (2, 3), (3, 1), (4, 5), (6, 7)])

# Find connected components
connected_components = list(nx.connected_components(G))

# Visualize the graph with connected components
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G, seed=42)  # Position nodes using the spring layout algorithm
colors = ['r', 'g', 'b', 'c', 'm', 'y', 'k']
for i, component in enumerate(connected_components):
    nx.draw_networkx_nodes(G, pos, nodelist=component, node_color=colors[i], label=f'Component {i+1}')
    nx.draw_networkx_edges(G, pos, width=1.0, alpha=0.5)
nx.draw_networkx_labels(G, pos, font_size=12, font_color='black', font_weight='bold')
plt.legend()
plt.axis('off')
plt.show()
```

In other words, an Eulerian circuit is a closed walk that includes all edges of the graph and visits each vertex exactly once, except for the starting vertex, which is visited twice (at the beginning and end of the circuit).

$\textbf{Characteristic}$

- Traverses Every Edge Exactly Once: An Eulerian circuit traverses each edge of the graph exactly once.
- Returns to Starting Vertex: It starts and ends at the same vertex, forming a closed loop.
- Visits Each Vertex Exactly Once (Except Starting Vertex): The circuit visits every vertex of the graph exactly once, except for the starting vertex, which is visited twice (at the beginning and end of the circuit).
- Even Degree Vertex: In a connected graph, every vertex has an even degree (even number of edges incident to it). In a graph with multiple connected components, exactly two vertices may have odd degrees, and they serve as the starting and ending points of the circuit.

$\textbf{Euler's Path}$

An Eulerian trail, also known as an Eulerian path, is a trail in a graph that traverses every edge exactly once. Unlike an Eulerian circuit, an Eulerian trail does not necessarily return to the starting vertex. However, it still visits every edge of the graph exactly once.

$\textbf{Characteristic:}$

- Traverses Every Edge Exactly Once: An Eulerian trail traverses each edge of the graph exactly once.
- Does Not Necessarily Return to Starting Vertex: Unlike an Eulerian circuit, an Eulerian trail does not necessarily return to the starting vertex. It starts at one vertex and ends at another, visiting every edge in between.
