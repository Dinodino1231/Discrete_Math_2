Task 1: Consider the following experiment:

1. The experiment consists of $n$ identical trials.

2. The outcome of each trial falls into one of $k$ classes or cells.

3. The probability that the outcome of a single trial falls into cell $i$, is $p_i,i = 1,2,...,k$ and remains the same from trial to trial. Notice that $p_1+p_2+p_3+···+p_k =1.$

4. The trials are independent.

5. The random variables of interest are $Y_1 , Y_2 , . . . , Y_k$, where $Y_i$ equals the number of trials for which the outcome falls into cell $i$. Notice that $Y_1 +Y_2 +Y_3 +···+Y_k =n$.

6. The joint probability distribution $P(y) = \frac{n!}{y_1!y_2!y_3!...y_n!} P_1^{y_1} P_2^{y_2} P_3^{y_3}....P_n^{y_n}$

Question 1: Show that the expected value $E[y] = \Sigma y \cdot P(y) = n \cdot p_i$

Question 2: Given the following data

\begin{gather}
\begin{bmatrix}
\begin{array}{c|c}
Age & Proportion \\
18-24 & 0.18 \\
25-34 & 0.23 \\
35-44 & 0.16 \\
45-64 & 0.27 \\
65-100 & 0.16 \\
\end{array}
\end{bmatrix}
\end{gather}

If 500 adults are sampled randomly, find the probability that the sample contains 100 persons between 18 and 24, 200 between the ages of 25 and 34, and 200 between the ages of 45 and 64. What is the expected value to obtain a person aged 65 and above?


**Solution**:

Question 1:

To show that the expected value $E[y] = \sum y \cdot P(y) = n \cdot p_i$, we need to calculate the expected value of the random variable $Y_i$, which represents the number of trials for which the outcome falls into cell $i$.

The expected value of $Y_i$, denoted as $E[Y_i]$, is given by:

$E[Y_i] = \sum_{y=0}^{n} y \cdot P(Y_i = y)$

Now, let's compute $P(Y_i = y)$, which is the probability that the number of trials for which the outcome falls into cell $i$ is $y$. According to the joint probability distribution:

$P(Y_i = y) = \frac{n!}{y!(n-y)!} p_i^y (1-p_i)^{n-y}$

Plugging this into the expression for $E[Y_i]$, we get:

$E[Y_i] = \sum_{y=0}^{n} y \cdot \frac{n!}{y!(n-y)!} p_i^y (1-p_i)^{n-y}$

To evaluate this sum, we can use the binomial expansion of $(p_i + (1-p_i))^n$, which equals $1^n = 1$. Thus, the sum becomes:

$E[Y_i] = \sum_{y=0}^{n} y \cdot \frac{n!}{y!(n-y)!} p_i^y (1-p_i)^{n-y} = n \cdot p_i$

Therefore, $E[Y_i] = n \cdot p_i$, which proves the desired result.




Question 2:

To find the probability that the sample contains 100 persons between 18 and 24, 200 between the ages of 25 and 34, and 200 between the ages of 45 and 64, we can use the multinomial probability formula:

$$
P(y) = \frac{n!}{y_1!y_2!y_3!} P_1^{y_1} P_2^{y_2} P_3^{y_3}
$$

Where $ n = 500 $, $ y_1 = 100 $, $ y_2 = 200 $, $ y_3 = 200 $, $ P_1 = 0.18 $, $ P_2 = 0.23 $, and $ P_3 = 0.27 $.

$$
P(y) = \frac{500!}{100!200!200!} (0.18)^{100} (0.23)^{200} (0.27)^{200}
$$

Then, to find the expected value to obtain a person aged 65 and above, we multiply the total sample size by the proportion of people aged 65 and above:

$$
\text{Expected Value} = 500 \times 0.16
$$

**Solution:**

**Question 1:**

Let $Y$ represent the number of successes observed during the $n$ trials.

The probability distribution of $Y$ follows a binomial distribution, denoted as $P(Y=y)$, where $y$ represents the number of successes.

The probability of observing exactly $y$ successes in $n$ trials can be calculated using the binomial probability formula:

$$P(Y=y) = \binom{n}{y} \cdot p^y \cdot (1-p)^{n-y}$$

where:
- $\binom{n}{y}$ represents the number of ways to choose $y$ successes out of $n$ trials, known as the binomial coefficient.
- $p$ is the probability of success on a single trial.
- $(1-p)$ is the probability of failure on a single trial.

This formula encapsulates the likelihood of observing $y$ successes out of $n$ trials, considering the independence and identical nature of each trial.


**Question 2:**

To find the expectation of the probability distribution, denoted as $E[Y]$, we sum the products of each possible outcome, $y$, and its corresponding probability, $P(Y=y)$:

$$E[Y] = \sum_{y=0}^{n} y \cdot P(Y=y)$$

Substituting the binomial probability formula, we get:

$$E[Y] = \sum_{y=0}^{n} y \cdot \binom{n}{y} \cdot p^y \cdot (1-p)^{n-y}$$

Expanding and simplifying this expression leads to:

$$E[Y] = \sum_{y=0}^{n} y \cdot \frac{n!}{y!(n-y)!} \cdot p^y \cdot (1-p)^{n-y}$$

After further algebraic manipulation, this simplifies to:

$$E[Y] = np$$

Hence, the expectation of the probability distribution is $E[Y] = np$.


**Question 3:**

Given that 30% of all persons afflicted by a certain illness recover, and a drug company claims that its medication is worthless, we want to find the probability that at least nine out of ten people receiving the medication will recover.

Since the medication is claimed to be worthless, we assume the probability of recovery, $p$, remains at 30%, or $0.3$, as per the given information.

Using the binomial distribution formula, we can calculate the probability of at least nine recoveries out of ten trials:

$$P(Y \geq 9) = P(Y = 9) + P(Y = 10)$$

Substituting the binomial probability formula, we get:

$$P(Y \geq 9) = \binom{10}{9} \cdot 0.3^9 \cdot (1-0.3)^{10-9} + \binom{10}{10} \cdot 0.3^{10} \cdot (1-0.3)^{10-10}$$

After calculation, this yields the probability of observing at least nine recoveries out of ten people who received the medication.


**Question 4:**

Given that a lot of 5000 electrical fuses contains 5% defectives, we want to find the probability of observing at least one defective fuse in a sample of 5 fuses.

Let $p$ be the probability of a fuse being defective, which is 5%, or $0.05$, as per the given information. Therefore, the probability of a fuse being non-defective is $1 - p = 0.95$.

Using the complement rule, the probability of observing at least one defective fuse is equal to 1 minus the probability of observing no defective fuses in the sample:

$$P(\text{at least one defective}) = 1 - P(\text{none defective})$$

The probability of observing none defective fuses in a sample of 5 is given by the binomial distribution:

$$P(\text{none defective}) = \binom{5}{0} \cdot 0.95^5$$

Substituting the values into the equation:

$$P(\text{at least one defective}) = 1 - \binom{5}{0} \cdot 0.95^5$$

After calculation, this yields the probability of observing at least one defective fuse in the sample.
